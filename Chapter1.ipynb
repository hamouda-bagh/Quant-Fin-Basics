#**1.2. The law of large numbers** 

```python
numbers2 = np.random.exponential(1, 10000)
Sn = np.cumsum(numbers2) / np.arange(1,10001)
```

```python
plt.plot(Sn)
```

Sn converges to 1 as N increases. 1 is the expectation of the exponential distribution with parameter 1.

```python
def emp_mean(x, N) :
  return np.cumsum(x[:N], axis=0)[-1] / N
```

```python
x = emp_mean(np.random.exponential(1, (10000, 10000)), 10000)
n, bins, patches = plt.hist(x, 50, density=True)
```

```python
n, bins, patches = plt.hist(x, 50, density=True, cumulative=True)
```

```python
xx = emp_mean(np.random.exponential(1, (10000,10000)), 100)
n, bins, patches = plt.hist(xx, 50, density=True)
```

```python
n, bins, patches = plt.hist(x, 50, density=True, cumulative=True)
```

We notice that for N=10000 the values are closer to the expectation than for N=100.

#**1.3. The central limit theorem**

```python
numbers3 = np.random.exponential(1, (10000, 10000))
```

```python
def Yn(x, N):
  return (np.cumsum(x[:N], axis=0)[-1] - N) / np.sqrt(N)
```

```python
Y = Yn(numbers3, 100)
```

```python
Y
```

```python
print(len(Y))
```

```python
n, bins, patches = plt.hist(Y, 50, density=True)
```

```python
y = np.random.randn(10000)
n, bins, patches = plt.hist(y, 50, density=True)
```

The histograms are similar.
